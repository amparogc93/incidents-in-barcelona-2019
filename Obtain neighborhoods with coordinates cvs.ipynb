#### Obtain coordinates

```python
import geocoder # import geocoder
    
geolocator = Nominatim(user_agent="foursquare_agent")
coordinates_list = []

# loop until you get the coordinates
for neigh in df['Neighborhood']:
    location = geolocator.geocode(neigh + ', Barcelona')
    latitude = location.latitude
    longitude = location.longitude
    coordinates_list.append([(neigh, latitude, longitude)])
    print(neigh, latitude, longitude)

coordinates = pd.DataFrame([item for coordinates_list in coordinates_list for item in coordinates_list])
coordinates.columns = ['Neighborhood', 'Neighborhood Latitude', 'Neighborhood Longitude']
coordinates.head()
```

#### Add the coordinates to the existing data frame

```python
# Merge both data frames so all the neighborhoods have their coordinates
df = pd.merge(df, coordinates, how='left', on='Neighborhood')
df.head()
```
